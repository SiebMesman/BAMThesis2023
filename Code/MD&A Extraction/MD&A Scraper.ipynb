## MD&A EDGAR Archives Scraper

Below is code obtained from the following URL: https://github.com/rflugum/10K-MDA-Section

This is a GitHub repository made by Ryan Flugum for the academic paper: Do Tone Changes in Financial Statements Predict Acquisition Behavior?" by John Berns, Patty Bick, Ryan Flugum and Reza Houston. See their repository for a more detailed explanation of the process and the required files.

As the code was a little outdated (due to a new version of Python, and extra security measures of the SEC website) some changes have been made which will be listed below:
1. A user agent is specified at the beginning of the code to overcome an error message regarding the use of automated tools.
2. A 3 second delay is added at the end of the code as the SEC website has an access limit per second.
3. Some parenthesis and little adjustments are added to update the Python code so that it is compatible with the latest version of Python (as of February 2023).
4. Content Encoder is set to 'utf-8' to prevent (most of the) errors when encoding the files.

Note that some files still return an error when trying to encode them, and thus this respective line must be deleted from the 'download_list' file before continuing.

#### Important Information:

This code will only work if a .txt file called 'download_list' is in the filepath specified within the code below. An example of such a .txt file can be found in the previously mentioned GitHub repository.

To make sure that you have personalized the code for your research, make the following changes to the code:

1. Change the email as specified in 'headers'.
2. Change the filepath to the path where the 'download_list.txt' files is stored.

#### Output and next steps:

The code will create a .txt file for every single line that is included in the 'download_list.txt' file, thus make sure that you have a folder ready that can store a big amount of files.

In the additional notebook called: 'MD&A Dataframe', all txt. files are read and stored within a Pandas DataFrame and exported as a csv file for future cleaning and preprocessing.
